Divide the data into two parts and name them 'train_data' and 'test data'. The 'train_data' should consist of all the dates before 01/01/2015. The test_data should consist of all the dates from 01/01/2015 to 03/31/2019.

```python
training_data = '2015-01-01'
test_data = '2015-01-01'
end_date = '2019-03-31'
train_data = df[df['date'] < test_data]
testing_data = df[(df['date'] >= test_data) & (df['date'] <= end_date)]
print(train_data)
print(testing_data)
```
