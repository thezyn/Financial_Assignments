```python
import datetime
import copy
import random

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsRegressor
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
```

```python
import pandas as pd
df = pd.read_csv('/data_ml.csv')
df.head()
#Delete columns one by using a loop:
for col in ['R1M_Usd', 'R6M_Usd', 'R12M_Usd']:
  del df[col]
print(df.columns)
```

Split Training & Testing


```python
#Target Variable
target_variable = 'R3M_Usd'

#define features
features = df.columns.tolist()
features.remove(target_variable)
features.remove('date')

#standardize the data
scaler = StandardScaler()

#fit the scaler on the training data and transform
train_data[features] = scaler.fit_transform(train_data[features])

#transform the test data
testing_data[features] = scaler.transform(testing_data[features])

# Transform the testing data using the fitted scaler
testing_data[features] = scaler.transform(testing_data[features])

# Extract target variables
y_train = train_data[target_variable]
y_test = testing_data[target_variable]

print(train_data)
print(testing_data)

```

Linear regression and Print MSE and R-square for the training and test data.


```python
#Create & Train Linear Regression Model
model = LinearRegression()
model.fit(train_data[features], y_train)

#Make predictions on training and testing data
y_train_pred = model.predict(train_data[features])
y_test_pred = model.predict(testing_data[features])

#Calculate MSE and R-Sqaured for training data
mse_train = mean_squared_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)

#Calculate MSE and R-Sqaured for testing data
mse_test = mean_squared_error(y_test, y_test_pred)
r2_test = r2_score(y_test, y_test_pred)

#Print the outcome
print("Training Data:")
print("MSE: ", mse_train)
print("R-Squared: ", r2_train)
print("\nTesting Data:")
print("MSE: ", mse_test)
print("R-Squared: ", r2_test)
```

Decision Tree and print the decision tree.


```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor, export_graphviz
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import graphviz

# Load the dataset
df = pd.read_csv('/data_ml.csv')  # Ensure the dataset is correctly loaded

# Define target variable and features
target_variable = 'R3M_Usd'
features = df.columns.tolist()
features.remove(target_variable)
features.remove('date')

# Split the data into training and testing sets
train_data, testing_data = train_test_split(df, test_size=0.2, random_state=42)

# Initialize and train the model with a restricted max depth for speed
model = DecisionTreeRegressor(max_depth=10, random_state=42)  # Limit depth for faster performance
model.fit(train_data[features], train_data[target_variable])

# Evaluate the model
predictions = model.predict(testing_data[features])
mse = mean_squared_error(testing_data[target_variable], predictions)
r2 = r2_score(testing_data[target_variable], predictions)
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")

# Visualize the Decision Tree (optional, skip if not needed)
dot_data = export_graphviz(model, out_file=None,
                           feature_names=features,
                           filled=True, rounded=True,
                           special_characters=True)
graph = graphviz.Source(dot_data)
# Only render graph when necessary to save time
# graph.render("decision_tree")

```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Plot the decision tree using sklearn's plot_tree
plt.figure(figsize=(20,10))  # Set the figure size
plot_tree(model, feature_names=features, filled=True, rounded=True)
plt.title("Decision Tree Visualization")
plt.show()  # This will display the plot inline in Google Colab
```
